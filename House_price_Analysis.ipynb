# Question - 4 

```python
import pandas as pd

# Convert the 'Price' column to numeric, if it's not already
prices_df['Price'] = pd.to_numeric(prices_df['price'], errors='coerce')

# Convert 'Price' based on 'Price Unit' (L = Lakhs, Cr = Crores)
prices_df['Price_in_numeric'] = prices_df.apply(
    lambda row: row['Price'] * 100 if row['price_unit'] == 'Cr' else row['Price'] * 1,
    axis=1)

# Find the house with the highest price
highest_price_house = prices_df.loc[prices_df['Price_in_numeric'].idxmax()]
```

```python
highest_price_house
```

```python
# Filter for houses with more than 3 BHK
houses_above_3_bhk = prices_df[prices_df['bhk'] > 3]

# Calculate the average area for those houses
average_area = houses_above_3_bhk['area'].mean()

# Display the average area
print(f"Average area of houses with more than 3 BHK: {average_area}")
```
